```python
# Checking bbox files
labels_root = "../data/yolo_sample/labels"
images_root = "../data/yolo_sample/images"
K = 6 

# collect all txts
txts = glob.glob(os.path.join(yolo_labels_root, "**", "*.txt"), recursive=True)
print(f"Found {len(txts)} label files")

random.seed(42)
sample = random.sample(txts, min(K, len(txts)))

def read_yolo_line(txt_path):
    with open(txt_path, "r") as f:
        for ln in f:
            ln = ln.strip()
            if not ln:
                continue
            parts = ln.split()
            if len(parts) != 5:
                return None
            cls = int(float(parts[0]))
            xc, yc, w, h = map(float, parts[1:])
            return cls, xc, yc, w, h
    return None

for t in sample:
    rel = t.split("/label3/", 1)[-1]                
    img_path = os.path.join(images_root, os.path.splitext(rel)[0] + ".jpg")

    if not os.path.exists(img_path):
        print(f"Image missing for {rel}")
        continue

    parsed = read_yolo_line(t)
    if parsed is None:
        print(f"No valid YOLO line in {t}")
        continue

    cls, xc, yc, w, h = parsed

    # load image to convert normalized coords → pixels
    img_bgr = cv2.imread(img_path)
    if img_bgr is None:
        print(f"Cannot read {img_path}")
        continue
    H, W = img_bgr.shape[:2]
    img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    # denormalize and get xyxy
    bw, bh = w * W, h * H
    x1 = int((xc * W) - bw / 2)
    y1 = int((yc * H) - bh / 2)
    x2 = int(x1 + bw)
    y2 = int(y1 + bh)

    # clamp to image
    x1 = max(0, min(W - 1, x1)); x2 = max(0, min(W - 1, x2))
    y1 = max(0, min(H - 1, y1)); y2 = max(0, min(H - 1, y2))

    # plot
    fig, ax = plt.subplots(1, 1, figsize=(7, 6))
    ax.imshow(img)
    rect = patches.Rectangle((x1, y1), x2 - x1, y2 - y1,
                             linewidth=2, edgecolor="red", facecolor="none")
    ax.add_patch(rect)
    ax.set_title(rel + f"  (cls={cls})")
    ax.axis("off")
    plt.show()
```

```python
from pathlib import Path
from PIL import Image
import shutil
from typing import Dict, Union

def save_crops_preserve_structure(
    images_root: Union[str, Path],
    labels_root: Union[str, Path],
    output_images_root: Union[str, Path],
    splits=("train","val","test"),
    pad_ratio: float = 0.02,
    min_crop_side: int = 8,
    clear_existing: bool = False,
    verbose: bool = True,
) -> Dict[str, int]:
    """
    Reads .txt files (3 lines: ignore 1+2, 3rd = pixel bbox x y w h or x1 y1 x2 y2),
    crops corresponding .jpg image, and writes to output preserving the
    original subfolder structure under each split.

    Example:
        images/train/123/45/2018/img001.jpg
        labels/train/123/45/2018/img001.txt
    →  output/train/123/45/2018/img001_crop.jpg
    """
    images_root = Path(images_root).resolve()
    labels_root = Path(labels_root).resolve()
    out_root    = Path(output_images_root).resolve()

    if clear_existing and out_root.exists():
        shutil.rmtree(out_root)
    out_root.mkdir(parents=True, exist_ok=True)

    def to_xyxy(a, b, c, d, W, H):
        if c > 0 and d > 0 and (a + c) <= (W + 2) and (b + d) <= (H + 2):
            x1, y1, x2, y2 = a, b, a + c, b + d
        else:
            x1, y1, x2, y2 = a, b, c, d
        pad = pad_ratio * max(abs(x2 - x1), abs(y2 - y1))
        x1 -= pad; y1 -= pad; x2 += pad; y2 += pad
        x1 = max(0, min(W - 1, int(round(x1))))
        y1 = max(0, min(H - 1, int(round(y1))))
        x2 = max(0, min(W,     int(round(x2))))
        y2 = max(0, min(H,     int(round(y2))))
        if x2 <= x1 or y2 <= y1: 
            return None
        if (x2 - x1) < min_crop_side or (y2 - y1) < min_crop_side:
            return None
        return x1, y1, x2, y2

    summary = {s: 0 for s in splits}

    for s in splits:
        lbl_base = labels_root / s
        txts = list(lbl_base.rglob("*.txt"))
        if verbose:
            print(f"[{s}] found {len(txts)} label files")

        for txt in txts:
            rel = txt.relative_to(lbl_base)
            imgp = (images_root / s / rel).with_suffix(".jpg")
            if not imgp.exists():
                continue

            try:
                lines = [ln.strip() for ln in txt.read_text(encoding="utf-8").splitlines() if ln.strip()]
                if len(lines) < 3:
                    continue
                nums = list(map(float, lines[2].split()))
                if len(nums) != 4:
                    continue
            except Exception:
                continue

            try:
                im = Image.open(imgp).convert("RGB")
            except Exception:
                continue
            W, H = im.size

            xyxy = to_xyxy(nums[0], nums[1], nums[2], nums[3], W, H)
            if xyxy is None:
                continue

            x1, y1, x2, y2 = xyxy
            crop = im.crop((x1, y1, x2, y2))

            # preserve subfolder structure
            outp = (out_root / s / rel).with_suffix(".jpg")
            outp.parent.mkdir(parents=True, exist_ok=True)

            try:
                crop.save(outp, format="JPEG", quality=95, subsampling=1, optimize=True)
                summary[s] += 1
            except Exception:
                continue

        if verbose:
            print(f"[{s}] crops saved: {summary[s]}")

    return summary
```

```python
from pathlib import Path
import os, tempfile, shutil, subprocess

ROOT = Path("../data/yolo_sample/crop_images").resolve()  # change
SPLITS = ["train", "val","test"]  

def walk_index():
    # returns {split: {make: {model: [img_paths...]}}}
    idx = {s: {} for s in SPLITS}
    for split in SPLITS:
        sdir = ROOT / split
        if not sdir.exists(): 
            continue
        for make_dir in sorted(d for d in sdir.iterdir() if d.is_dir()):
            make = make_dir.name
            idx[split].setdefault(make, {})
            for model_dir in sorted(d for d in make_dir.iterdir() if d.is_dir()):
                model = model_dir.name
                idx[split][make].setdefault(model, [])
                for year_dir in sorted(d for d in model_dir.iterdir() if d.is_dir()):
                    for img in year_dir.rglob("*"):
                        if img.suffix.lower() in {".jpg"} and img.is_file():
                            idx[split][make][model].append(img)
    return idx
```

```python
def ln(src: Path, dst: Path):
    dst.parent.mkdir(parents=True, exist_ok=True)
    try:
        os.symlink(src, dst)
    except FileExistsError:
        pass

def make_temp_dataset_for_make(idx):
    tmp = Path(tempfile.mkdtemp(prefix="yolo_make_"))
    for split in SPLITS:
        for make, models in idx.get(split, {}).items():
            for model, imgs in models.items():
                for img in imgs:
                    # flatten to .../<tmp>/<split>/<make>/<filename>
                    ln(img, tmp / split / make / img.name)
    return tmp  # caller should clean it

def make_temp_dataset_for_models_of(make, idx):
    tmp = Path(tempfile.mkdtemp(prefix=f"yolo_model_{make}_"))
    for split in SPLITS:
        models = idx.get(split, {}).get(make, {})
        for model, imgs in models.items():
            for img in imgs:
                # .../<tmp>/<split>/<model>/<filename>
                ln(img, tmp / split / model / img.name)
    return tmp
```

```python
import subprocess
def run_yolo_classify_train(yolo_exe, data_dir, project_dir, run_name, device, epochs=100, imgsz=224, model="yolo11n-cls.pt",  extra_overrides=None,):
    cmd = [
        str(yolo_exe), "classify", "train",
        f"model={model}",
        f"data={str(data_dir)}",
        f"epochs={epochs}",
        f"imgsz={imgsz}",
        f"project={str(project_dir)}",
        f"name={run_name}",
        f"device={device}",
    ]
    if extra_overrides:
        for k, v in extra_overrides.items():
            cmd.append(f"{k}={v}")
    print("Running:", " ".join(cmd))
    subprocess.run(cmd, check=True)
```

```python
for make in sorted({m for s in SPLITS for m in idx.get(s, {}).keys()}):
    if str(make) not in allowed:
        print(f"Skipping make {make} (not in top15)")
        continue

    tmp_models = make_temp_dataset_for_models_of(make, idx)
    tmp_model_top15 = tmp_models.parent / (tmp_models.name + "_top15")
    tmp_model_top15.mkdir(parents=True, exist_ok=True)

    # Copy only allowed make’s models into tmp_model_top15
    copied = False
    for split in SPLITS:
        src_split = tmp_models / split
        dst_split = tmp_model_top15 / split
        if not src_split.exists():
            continue

        for model_dir in src_split.iterdir():
            # Each folder here is a model name under this make
            shutil.copytree(model_dir, dst_split / model_dir.name, dirs_exist_ok=True)
            copied = True

    if not copied:
        print(f"⚠️ No data copied for make {make}, skipping training.")
        shutil.rmtree(tmp_models, ignore_errors=True)
        shutil.rmtree(tmp_model_top15, ignore_errors=True)
        continue

    try:
        run_yolo_classify_train(
            yolo_executable,
            tmp_model_top15,
            project_path,
            run_name=f"model_cls_{make}_y11n_100e",
            device="mps",  # remove if not using Apple MPS
            extra_overrides={"amp": "False"}  # Key line for MPS compatibility
        )
    finally:
        shutil.rmtree(tmp_models, ignore_errors=True)
        shutil.rmtree(tmp_model_top15, ignore_errors=True)
```

```python
# IDs match folder naming (strings vs ints)
id2name = dict(zip(df_make_id["make_id"].astype(str), df_make_id["make_names"].astype(str)))

def to_make_name(x):
    # x might already be a name; fall back gracefully
    return id2name.get(str(x), str(x))
```

```python
import os, glob
import pandas as pd
from ultralytics import YOLO

test_data_path = '../data/yolo_sample/crop_images/test'
model_path = '../data/yolo_sample/runs/classify_top15/make_cls_y11n_100e2/weights/best.pt'  # set this

model = YOLO(model_path)

allowed_ids = set(df_make_top15["make_id"].astype(str)) 

true_labels = []
predicted_labels = []
predicted_probs = []
image_paths = []

# discover classes (these might be numeric folder names like "100", "101", ...)
class_ids = sorted([d for d in os.listdir(test_data_path) 
                    if os.path.isdir(os.path.join(test_data_path, d))])

# filter only those classes that are in top15
class_ids = [cid for cid in class_ids if cid in allowed_ids]

print(f'Found {len(class_ids)} classes (IDs): {class_ids[:10]}...')

for class_id in class_ids:
    class_dir = os.path.join(test_data_path, class_id)
    # recursive in case you have nested folders
    image_files = glob.glob(os.path.join(class_dir, '**', '*.*'), recursive=True)
    # keep only image extensions
    image_files = [p for p in image_files if os.path.splitext(p)[1].lower() in 
                   {'.jpg'}]
    print(f'Processing {class_id} ({to_make_name(class_id)}): {len(image_files)} images')

    for img_path in image_files:
        try:
            res = model.predict(source=img_path, imgsz=224, verbose=False)[0]
            pred_id = int(res.probs.top1)
            pred_name_from_dataset = res.names[pred_id]  # whatever your dataset folder advertised
            # normalize both sides to make names
            true_labels.append(to_make_name(class_id))
            predicted_labels.append(to_make_name(pred_name_from_dataset))
            predicted_probs.append(float(res.probs.top1conf))
            image_paths.append(img_path)
        except Exception as e:
            print(f'Error processing {img_path}: {e}')
            continue

print(f'\nEvaluation completed: {len(true_labels)} images processed')

# save a predictions CSV with make names
df_pred = pd.DataFrame({
    "image": image_paths,
    "true_make": true_labels,
    "pred_make": predicted_labels,
    "pred_conf": predicted_probs
})
out_csv = "./test_predictions_make_names.csv"
df_pred.to_csv(out_csv, index=False)
print(f"Saved per-image predictions → {out_csv}")
```

```python
import os, glob
import pandas as pd
from ultralytics import YOLO

# === Configuration ===
test_data_path = '../data/yolo_sample/crop_images/test'
runs_dir = '../data/yolo_sample/runs/classify_top15'
make_model_pattern = 'model_cls_{}_y11n_100e'     # submodel folder name pattern per make_id
make_model_weights = 'weights/best.pt'
make_cls_path = f'{runs_dir}/make_cls_y11n_100e2/weights/best.pt'

# === Mappings (you already have these) ===
id2name = {str(i): str(n) for i, n in zip(df_make_id["make_id"], df_make_id["make_names"])}
name2id = {str(n).lower(): str(i) for i, n in zip(df_make_id["make_names"], df_make_id["make_id"])}
allowed_ids = set(df_make_top15["make_id"].astype(str))

def to_make_name(x):
    return id2name.get(str(x), str(x))

def extract_true_model_id(img_path, true_make_id):
    """Expect: test/<make_id>/<model_id>/.../image.jpg → return model_id (str)"""
    rel = os.path.relpath(img_path, start=os.path.join(test_data_path, str(true_make_id)))
    parts = rel.split(os.sep)
    return parts[0] if len(parts) >= 2 else None  # first folder under make_id

# === Load make classifier ===
make_model = YOLO(make_cls_path)
print(f"✅ Loaded make classifier from: {make_cls_path}")

# === Cache per-make model classifiers ===
model_cache = {}

records = []

# === Discover allowed makes in test set ===
class_ids = sorted([d for d in os.listdir(test_data_path)
                    if os.path.isdir(os.path.join(test_data_path, d)) and d in allowed_ids])
print(f'Found {len(class_ids)} allowed make folders.')

for class_id in class_ids:
    class_dir = os.path.join(test_data_path, class_id)
    image_files = glob.glob(os.path.join(class_dir, '**', '*.jpg'), recursive=True)

    for img_path in image_files:
        try:
            # ---- true labels (make + model) ----
            true_make_name = to_make_name(class_id)
            true_model_id = extract_true_model_id(img_path, class_id)
            if true_model_id is None:
                # Skip images not under a model folder (shouldn't happen in your data)
                continue
            true_model = str(true_model_id)  # map to name if you have df_model_id

            # ---- Stage 1: predict make ----
            make_res = make_model.predict(source=img_path, imgsz=224, verbose=False)[0]
            pred_idx = int(make_res.probs.top1)
            pred_label_str = make_res.names[pred_idx]              # e.g., "97" (string id) or a name
            pred_make_name = id2name.get(str(pred_label_str), str(pred_label_str))
            pred_make_id = pred_label_str
            pred_make_conf = float(make_res.probs.top1conf)

            # ---- Stage 2: predict model within predicted make ----
            submodel_dir = os.path.join(runs_dir, make_model_pattern.format(pred_make_id))
            weights_path = os.path.join(submodel_dir, make_model_weights)

            pred_model = None
            pred_model_conf = None

            if os.path.exists(weights_path):
                if pred_make_id not in model_cache:
                    model_cache[pred_make_id] = YOLO(weights_path)
                model_cls = model_cache[pred_make_id]

                model_res = model_cls.predict(source=img_path, imgsz=224, verbose=False)[0]
                top1_id = int(model_res.probs.top1)
                pred_model = model_res.names[top1_id]              # model label string (e.g., "874")
                pred_model_conf = float(model_res.probs.top1conf)

            records.append({
                "image": img_path,
                "true_make": true_make_name,
                "true_model": true_model,
                "pred_make": pred_make_name,
                "pred_make_conf": pred_make_conf,
                "pred_model": pred_model,
                "pred_model_conf": pred_model_conf
            })

        except Exception as e:
            print(f"Error on {img_path}: {e}")
            continue

# === Save all predictions ===
df_all = pd.DataFrame(records)
out_csv = "./test_predictions_make_model.csv"
df_all.to_csv(out_csv, index=False)
print(f"Saved combined predictions → {out_csv}")

# === MODEL-LEVEL METRICS ===
if df_all.empty:
    print("No predictions generated.")
else:
    model_known = df_all.dropna(subset=["pred_model"]).copy()
    print(f"\nImages processed: {len(df_all)}")
    print(f"Images with model predictions: {len(model_known)}")

    if model_known.empty:
        print("⚠️ No model-level predictions available (check trained submodels and names).")
    else:
        # Compare true_model vs pred_model (both strings)
        acc_model_overall = (model_known["true_model"] == model_known["pred_model"]).mean()
        print(f"Overall Model Accuracy: {acc_model_overall:.3f}")

        # Per-make model accuracy
        per_make = (model_known.assign(correct=(model_known["true_model"] == model_known["pred_model"]))
                              .groupby("true_make")["correct"]
                              .mean()
                              .sort_values(ascending=False))
        print("\nPer-make Model Accuracy:")
        print(per_make.to_string())

        # (optional) Save per-make summary
        per_make.to_frame("model_accuracy").to_csv("./model_accuracy_per_make.csv")
        print("Saved per-make model accuracy → ./model_accuracy_per_make.csv")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score

# Expect df_all to have: ["true_make","true_model","pred_model"]
# Keep rows where we have both true & predicted model labels
valid = df_all.dropna(subset=["true_model", "pred_model"]).copy()

# Normalize labels to avoid casing/whitespace mismatches
valid["true_model_norm"] = valid["true_model"].astype(str).str.strip().str.lower()
valid["pred_model_norm"] = valid["pred_model"].astype(str).str.strip().str.lower()

# Compute per-make aggregated model F1
rows = []
for make in sorted(valid["true_make"].unique()):
    sub = valid[valid["true_make"] == make]
    y_true = sub["true_model_norm"].tolist()
    y_pred = sub["pred_model_norm"].tolist()
    # Use union of labels seen to avoid undefined metrics; zero_division=0 for safety
    labels = sorted(list(set(y_true) | set(y_pred)))
    f1_macro = f1_score(y_true, y_pred, labels=labels, average="macro", zero_division=0)
    f1_weighted = f1_score(y_true, y_pred, labels=labels, average="weighted", zero_division=0)
    rows.append({"make": make, "f1_macro": f1_macro, "f1_weighted": f1_weighted, "support": len(sub)})

make_f1 = pd.DataFrame(rows).sort_values("f1_macro", ascending=False).reset_index(drop=True)
print(make_f1)

# Top/Bottom makes by model F1 (macro)
print("\n" + "="*60)
print("TOP 5 MAKES BY AGGREGATED MODEL F1 (macro)")
print("="*60)
for i, r in make_f1.head(5).iterrows():
    print(f"{i+1:2d}. {r['make']:12} | F1: {r['f1_macro']:.3f} | Support: {int(r['support'])}")

print("\n" + "="*60)
print("BOTTOM 5 MAKES BY AGGREGATED MODEL F1 (macro)")
print("="*60)
for i, r in make_f1.tail(5).iterrows():
    print(f"{len(make_f1)-4+i:2d}. {r['make']:12} | F1: {r['f1_macro']:.3f} | Support: {int(r['support'])}")

# Bar plot (macro F1) – one bar per make
plt.figure(figsize=(14, 7))
classes = make_f1["make"].tolist()
f1_scores = make_f1["f1_macro"].tolist()

# Color by threshold
colors = [
    "green" if f1 > 0.8 else ("orange" if f1 > 0.6 else "red")
    for f1 in f1_scores
]

bars = plt.bar(range(len(classes)), f1_scores, color=colors)
plt.title("Aggregated Car-Model F1 by Car Make (macro)", fontsize=16, pad=20)
plt.xlabel("Car Make", fontsize=12)
plt.ylabel("F1-Score (macro)", fontsize=12)
plt.xticks(range(len(classes)), classes, rotation=45, ha="right")
plt.ylim(0, 1.0)
plt.grid(axis="y", alpha=0.3)

# Add value labels
for bar, f1 in zip(bars, f1_scores):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f"{f1:.2f}", ha="center", va="bottom", fontsize=8)

plt.tight_layout()

# Choose where to save
out_dir = "../data/yolo_sample/runs/classify_top15"
os.makedirs(out_dir, exist_ok=True)
f1_plot_path = os.path.join(out_dir, "agg_model_f1_by_make.png")
plt.savefig(f1_plot_path, dpi=300, bbox_inches="tight")
plt.show()
print(f"\nF1 plot saved to: {f1_plot_path}")

# (Optional) Save the table too
make_f1.to_csv(os.path.join(out_dir, "agg_model_f1_by_make.csv"), index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report

# Suppose your dataframe (df_all or model_known) has:
#   'true_model_norm', 'pred_model_norm', 'model_conf', 'image' columns
y_true = model_known['true_model_norm'].tolist()
y_pred = model_known['pred_model_norm'].tolist()
confidences = model_known['pred_model_conf'].tolist() if 'pred_model_conf' in model_known else [np.nan]*len(y_true)
image_paths = model_known['image'].tolist() if 'image' in model_known else list(range(len(y_true)))

# --- Create detailed results ---
results_df = pd.DataFrame({
    'image_path': image_paths,
    'true_label': y_true,
    'predicted_label': y_pred,
    'confidence': confidences,
    'correct': [t == p for t, p in zip(y_true, y_pred)]
})

# --- Save detailed results ---
out_path = './detailed_model_predictions.csv'
results_df.to_csv(out_path, index=False)
print(f"Detailed model predictions saved to: {out_path}")

# --- Misclassifications ---
misclassified = results_df[~results_df['correct']]
if len(misclassified) > 0:
    print(f'\nFound {len(misclassified)} misclassified images')
    print('Sample misclassifications:')
    print(misclassified[['true_label', 'predicted_label', 'confidence']].head(10))
else:
    print('\nNo misclassifications found — perfect accuracy!')

# --- Summary Statistics ---
total = len(results_df)
correct = results_df['correct'].sum()
accuracy = correct / total if total > 0 else 0.0
avg_conf = np.mean(confidences)
std_conf = np.std(confidences)

print(f'\n{"="*50}')
print('SUMMARY STATISTICS (Model-level)')
print(f'{"="*50}')
print(f'Total images evaluated: {total}')
print(f'Correctly classified: {correct}')
print(f'Misclassified: {total - correct}')
print(f'Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)')
print(f'Average confidence: {avg_conf:.3f}')
print(f'Confidence std: {std_conf:.3f}')

# --- Optional: save classification report ---
labels = sorted(list(set(y_true) | set(y_pred)))
report = classification_report(y_true, y_pred, labels=labels, output_dict=True, zero_division=0)
report_df = pd.DataFrame(report).T
report_df.to_csv('./per_model_classification_report.csv', index=True)
print(f"\nDetailed per-model classification report saved → ./per_model_classification_report.csv")
```
